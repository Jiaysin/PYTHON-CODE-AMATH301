```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[-11,45,-33],[7,2,-25],[0,-8,4]]) #(1)

#A = np.array([[-11,45,-33],[7,2,-25],[0,-8,4],[1,1,1]]) #(2)
```

```python
#(a)

Ub,Sb,VTb=np.linalg.svd(A)
print('U =',Ub)
print('Sigma (vector) =',Sb)
print('V^T =',VTb)
```

```python
#(b)

Smat = np.zeros(np.shape(A))
Smat[0:np.min(np.shape(A)),0:np.min(np.shape(A))] = np.diag(Sb)
print('Sigma (matrix) =',Smat)
print('A =?',Ub@Smat@VTb)
```

```python
#(c)

eigvalATA, eigvecATA = np.linalg.eig(A.T@A)
eigvalAAT, eigvecAAT = np.linalg.eig(A@A.T)
print('eigvals of A^T A =',eigvalATA)
print('eigvals of A A^T =',eigvalAAT)
print('singular values of A =',np.sqrt(eigvalAAT))
```

```python
#(d)

print('eigvecs of A A^T =',eigvecAAT)
print('eigvecs of A^T A =',eigvecATA)
```

```python
#(3b)

A = np.array([[0.98,1.,0.99,0.02,0.01,0.03,0.05,0.98],
[0.99,0.97,0.96,0.01,0.04,0.02,0.01,0.99],
[0.96,0.99,0.92,0.05,0.02,0.02,0.02,0.96],
[0.02,0.01,0.03,0.91,0.98,0.06,0.01,0.96],
[0.03,0.08,0.04,0.99,0.94,0.01,0.04,0.99],
[0.,0.02,0.05,0.96,0.98,0.02,0.04,0.99]])
print(A)

U,S,VT=np.linalg.svd(A)
print('S =',S)
```

```python
U2 = U[:,0:2]
S2 = np.diag(S[0:2])
VT2 = VT[0:2,:]

B = U2@S2
print('B =',B)

C = S2@VT2
print('C =',C)
```

```python
plt.plot(B[:,0],B[:,1],'ro')
labels = np.array(['y1','y2','y3','n1','n2','n3'])
for i in range(0,len(labels)):
    plt.annotate(labels[i], xy=(B[i,0], B[i,1]))
```

```python
plt.plot(C[0,:],C[1,:],'ro')
labels = np.array(['gene1','gene2','gene3','gene4','gene5','gene6','gene7','gene8'])
for i in range(0,len(labels)):
    plt.annotate(labels[i], xy=(C[0,i], C[1,i]))
```
