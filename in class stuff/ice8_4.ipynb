```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
#1
def f(t,v): #v' = f(t,v) where v=[y,z]^T
    output = np.array([v[1],np.exp(t)-2*v[1]-2*v[0]])
    return output

#SOLVE NUMERICALLY WITH SOLVE_IVP-------------------------
tspan = np.array([0,1]) #starting and ending time
tt = np.linspace(tspan[0],tspan[1],1001)

alpha = 0

ic = np.array([2,alpha])
sol = solve_ivp(f,tspan,ic,method="RK45",t_eval=tt)
ysolve = sol.y[0]
plt.plot(tt,ysolve,'-m') #y(t)
yfin = ysolve[-1]
print(yfin)

#PLOT TRUE SOLUTION
#alpha = 5.179188
#ic = np.array([2,alpha])
#sol = solve_ivp(f,tspan,ic,method="RK45",t_eval=tt)
#ysolve = sol.y[0]
#plt.plot(tt,ysolve,'-g') #y(t)
```

```python
#2
def f(t,v): #v' = f(t,v) where v=[y,z]^T
    output = np.array([v[1],2*v[1]-t**2*v[0]])
    return output

#SOLVE NUMERICALLY WITH SOLVE_IVP-------------------------
tspan = np.array([0,2]) #starting and ending time
tt = np.linspace(tspan[0],tspan[1],1001)

#(a)
alpha = -5
#alpha = 5
ic = np.array([3,alpha])
sol = solve_ivp(f,tspan,ic,method="RK45",t_eval=tt)
ysolve = sol.y[0]
plt.plot(tt,ysolve,'-m') #y(t)
yfin = ysolve[-1]
print(yfin) #(b) g(alpha)
```

```python
#2(c)

tspan = np.array([0,5]) #starting and ending time
tt = np.linspace(tspan[0],tspan[1],1001)

alphavec = np.arange(-5,5.01,1)
for alpha in alphavec:
    ic = np.array([3,alpha])
    sol = solve_ivp(f,tspan,ic,method="RK45",t_eval=tt)
    ysolve = sol.y[0]
    plt.plot(tt,ysolve,'-m') #y(t)
```
