```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fminbound
```

```python
#1(a)

def f(x):
    return x+1/x

a = 0.5
b = 5
funceval = 0

for i in range(0,101):
    x1 = 2*a/3+b/3
    x2 = a/3+2*b/3
    print(a,x1,x2,b)

    if f(x1)<f(x2):
        b = x2
    else: #f(x1)>=f(x2), including the edge-case of equality
        a = x1
    funceval += 2
    
    if b-a<1e-6:
        break

finalvalue = (a+b)/2
print('final interval: [',a,',',b,']')
print('final value: ',finalvalue)
print('number of iterations (trisection): ',i)
print('number of function evaluations (trisection):',funceval)
```
