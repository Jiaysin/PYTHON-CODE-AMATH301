```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1

avec = np.array([7])

for n in range(1,100):
  an = avec[-1] #a_n = last entry in avec
  if np.mod(an,2)==0: #a_n is even
    anpo = an/2 #a_(n+1)=a_n/2
  else: #a_n is odd
    anpo = 3*an+1 #a_(n+1)=3a_n+1

  avec = np.append(avec,anpo)

  if anpo==1:
    break

print('N =',n)
print('a =',avec)
```

```python
#2

numiters = np.array([]) #number of iterations
startvals = np.arange(1,100) #starting values

for astart in startvals:
  #below is copied from problem 1---
  avec = np.array([astart])

  for n in range(1,100):
    an = avec[-1] #a_n = last entry in avec
    if np.mod(an,2)==0: #a_n is even
      anpo = an/2 #a_(n+1)=a_n/2
    else: #a_n is odd
      anpo = 3*an+1 #a_(n+1)=3a_n+1

    avec = np.append(avec,anpo)

    if anpo==1:
      break
  #above is copied from problem 1---

  numiters = np.append(numiters,n)

plt.plot(startvals,numiters,'.b')
```

```python
#3

highvals = np.array([]) #highest value reached
startvals = np.arange(1,100) #starting values

for astart in startvals:
  avec = np.array([astart])
  highsofar = astart #highest value reached so far

  for n in range(1,1000):
    an = avec[-1] #a_n = last entry in avec
    if np.mod(an,2)==0: #a_n is even
      anpo = an/2 #a_(n+1)=a_n/2
    else: #a_n is odd
      anpo = 3*an+1 #a_(n+1)=3a_n+1
      if anpo>highsofar:
        highsofar = anpo

    avec = np.append(avec,anpo)

    if anpo==1:
      break

  highvals = np.append(highvals,highsofar)

plt.plot(startvals,highvals,'.b')

#plt.plot(startvals,highvals,'.b',markersize=1)
#plt.axis([1, 10000, 1, 10000])
```
